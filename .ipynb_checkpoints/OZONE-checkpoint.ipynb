```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from sorted_months_weekdays import *

from sort_dataframeby_monthorweek import *
```

```python
df = pd.read_csv("./data/new_combined_ozone.csv")
```

```python
print(df.shape)
df.head()
```

```python
print(df['DAILY_AQI_VALUE'].skew())
df['DAILY_AQI_VALUE'].describe()
```

```python
print(df['Daily Max 8-hour Ozone Concentration'].skew())
df['Daily Max 8-hour Ozone Concentration'].describe()
```

```python
lower_quantile, upper_quantile = df['DAILY_AQI_VALUE'].quantile([.1, .9])
```

```python
df1=df.loc[(df['DAILY_AQI_VALUE'] > lower_quantile) & (df['DAILY_AQI_VALUE'] < upper_quantile)]
print(df1.shape)
print(df1['DAILY_AQI_VALUE'].skew())
df1['DAILY_AQI_VALUE'].describe()
```

```python
county_wise ={}
for c in df1['County'].unique():
    temp = df1[df1['County'] == c]
    
    county_wise[c] = temp
    fig = go.Figure([go.Scatter(x=temp['Date'], y=temp['DAILY_AQI_VALUE'])])
    fig.update_layout(
    title="DAILY_AQI_VALUE in " +c ,
    xaxis_title="Date",
    yaxis_title="DAILY AQI VALUE",
    legend_title= c,
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"))
    fig.show()
```

```python
for d in county_wise:
    county_wise[d]['Date'] = pd.to_datetime(county_wise[d]['Date'], errors='coerce')
    print("------------------------")
    print("For "+ d +" AVG by month")
    s= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['DAILY_AQI_VALUE'].sum()
    case= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['New Positives'].sum()
#     print(s)
    count = county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['DAILY_AQI_VALUE'].count()
#     print(count)

#     print(s/count)
    avg = pd.DataFrame(s/count)
    avg.reset_index(level=0, inplace=True)
    
    avg.columns = ["Month", " AVG DAILY_AQI_VALUE"]
    avg["Total DAILY_AQI_VALUE"] =s.values 
    avg["count"] = count.values
    avg["Total Cases"] = case.values
    avg["Avg Cases"] = (case/count).values
    
    avg =Sort_Dataframeby_Month(df=avg,monthcolumnname='Month')
    print(avg)
    print("------------------------")
```

```python
# without removing the outliers
# maybe it was the outliers which made sense for this. 

county_wise ={}
for c in df['County'].unique():
    temp = df[df['County'] == c]
    
    county_wise[c] = temp
    fig = go.Figure([go.Scatter(x=temp['Date'], y=temp['DAILY_AQI_VALUE'])])
    fig.update_layout(
    title="DAILY_AQI_VALUE in " +c ,
    xaxis_title="Date",
    yaxis_title="DAILY AQI VALUE",
    legend_title= c,
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"))
    fig.show()
```

```python
all_avg ={}
for d in county_wise:
    county_wise[d]['Date'] = pd.to_datetime(county_wise[d]['Date'], errors='coerce')
    print("------------------------")
    print("For "+ d +" AVG by month")
    s= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['DAILY_AQI_VALUE'].sum()
    case= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['New Positives'].sum()
    total_testing= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['Total Number of Tests Performed'].sum()
#     print(s)
    count = county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['DAILY_AQI_VALUE'].count()
#     print(count)

#     print(s/count)
    avg = pd.DataFrame(s/count)
    avg.reset_index(level=0, inplace=True)
    
    avg.columns = ["Month", "AVG DAILY_AQI_VALUE"]
    avg["Total DAILY_AQI_VALUE"] =s.values 
    avg["count"] = count.values
    avg["Total Cases"] = case.values
    avg["Avg Cases"] = (case/count).values
    avg["Avg Test"] = (total_testing/count).values
    avg["Percentage Positive"] = (avg["Avg Cases"]*100)/ avg["Avg Test"]
    
    avg =Sort_Dataframeby_Month(df=avg,monthcolumnname='Month')
    print(avg)
    all_avg[d] = avg
    print("------------------------")
```

```python
for d in all_avg:
    temp = all_avg[d]
    fig = px.bar(temp, x="Month", y="AVG DAILY_AQI_VALUE", hover_data=['Avg Cases', 'Avg Test', 'Percentage Positive'], color='AVG DAILY_AQI_VALUE',
                 labels={'Month':'Month wise distribution in '+d} ,height=400)
    fig.show()
```
