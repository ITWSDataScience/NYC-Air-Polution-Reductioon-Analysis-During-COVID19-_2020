```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data loading
covid_data = pd.read_csv("data/New_York_State_Statewide_COVID-19_Testing.csv")
covid_data.head(0)
```

```python
covid_data.describe()
```

```python
print(covid_data.isna().sum(), covid_data.dtypes)
```

```python
all_county=covid_data['County'].unique()
for county in all_county:
    covid_data.loc[covid_data['County'] == county].plot(x='Test Date', y= 'New Positives', label = county)
    plt.xlabel('Test Date')
    plt.ylabel("New Positives")
    plt.legend(loc="upper left")
    
```
